```python
# Download and read the `spacex_launch_geo.csv`
from js import fetch
import io

URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df=pd.read_csv(spacex_csv_file)
```

```python
# Task 2: Mark the success/failed launches for each site on the map
```

Next, let's try to enhance the map by adding the launch outcomes for each site, and see which sites have high success rates.
Recall that data frame spacex_df has detailed launch records, and the `class` column indicates if this launch was successful or not


```python
spacex_df.tail(10)
```

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
# Function to assign color to launch outcome
def assign_marker_color(launch_outcome):
    if launch_outcome == 1:
        return 'green'
    else:
        return 'red'

spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)
spacex_df.tail(10)
```

*TODO:* For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python
# Add marker_cluster to current site_map
site_map.add_child(marker_cluster)

# for each row in spacex_df data frame, create and add a Marker to the marker_cluster
for index, record in spacex_df.iterrows():
    # Get coordinate and color
    coordinate = [record['Lat'], record['Long']]
    color = record['marker_color']

    # Create a Marker object with its coordinate and color
    marker = folium.Marker(
        coordinate,
        icon=folium.Icon(color='white', icon_color=color), # Use icon_color for the colored pin
        popup=f"Launch: {index}, Class: {record['class']}"
    )
    # Add marker to the cluster
    marker_cluster.add_child(marker)

# Display the map (The map will be displayed in the notebook output)
site_map
```

Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

Based on the visualized proximities, here are the general findings for a typical launch site like KSC:

    Are launch sites in close proximity to railways?

        Yes. The closest railway is often very close (e.g., KSC is about 1 km away from a railway line). Railways are essential for transporting large components like rocket stages and propellants.

    Are launch sites in close proximity to highways?

        Yes. The closest highway/major road is often also very close (e.g., KSC is about 7-8 km away from a major route). Highways are necessary for general access and logistics for personnel and smaller components.

    Are launch sites in close proximity to the coastline?

        Yes. Launch sites are built right on the coast (e.g., KSC is about 7-8 km from the coastline). This proximity allows rockets to launch over the open ocean, maximizing safety by preventing debris from falling onto populated areas during launch or in case of a failure.

    Do launch sites keep a certain distance away from cities?

        Yes. Launch sites maintain a deliberate distance from major cities (e.g., KSC is about 15-20 km from Titusville, a relatively small city). This separation is a critical safety measure to protect the public from noise, pollution, and the risk of catastrophic failure.

The combination of being close to the equator (for efficiency), the coast (for safety), and transport infrastructure (for logistics), while maintaining a safe distance from major population centers, defines the optimal geographical location for a rocket launch site.
